```python
# /*==========================================================================================*\
# **                        _           _ _   _     _  _         _                            **
# **                       | |__  _   _/ | |_| |__ | || |  _ __ | |__                         **
# **                       | '_ \| | | | | __| '_ \| || |_| '_ \| '_ \                        **
# **                       | |_) | |_| | | |_| | | |__   _| | | | | | |                       **
# **                       |_.__/ \__,_|_|\__|_| |_|  |_| |_| |_|_| |_|                       **
# \*==========================================================================================*/


# -----------------------------------------------------------------------------------------------
# Author: Bùi Tiến Thành - Tien-Thanh Bui (@bu1th4nh)
# Title: upgrade_test_result_to_mongoDB.ipynb
# Date: 2024/12/09 13:53:16
# Description: 
# 
# (c) 2024 bu1th4nh. All rights reserved. 
# Written with dedication in the University of Central Florida, EPCOT and the Magic Kingdom.
# -----------------------------------------------------------------------------------------------


import logging
import numpy as np
import pandas as pd
from tqdm import tqdm
from typing import List, Dict, Any, Tuple, Union, Literal


import pymongo
import mlflow


mongo = pymongo.MongoClient(
    host='mongodb://localhost',
    port=27017,
    username='bu1th4nh',
    password='ariel.anna.elsa',
)
mongo_db = mongo['SimilarSampleCrossOmicNMF']
mlflow.set_tracking_uri(uri="http://127.0.0.1:6969")




configs = [
    ('SimilarSampleCrossOmicNMFv3', 'BRCA'),
    ('SimilarSampleCrossOmicNMFv3_LUAD', 'LUAD'),
    ('SimilarSampleCrossOmicNMFv3_OV', 'OV'),
]

for exp_name, collection in configs:
    mlflow.set_experiment(exp_name)
    Ariel = mongo_db[collection]  

    run_df = mlflow.search_runs()[['run_id', 'tags.mlflow.runName']]

    for idx in run_df.index:
        run_id = run_df.loc[idx, 'run_id']
        run_name = run_df.loc[idx, 'tags.mlflow.runName']
        
        if Ariel.find_one({'run_id': run_id}) is None: continue

        Ariel.update_many(
            {'run_id': run_id},
            {'$set': {'run_name': run_name}}
        )

```
